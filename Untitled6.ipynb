# Lets discuss about principle compound analysis (PCA)
* Principal Component Analysis (PCA) is a statistical technique used to simplify complex datasets. It does this by reducing the number of variables (or dimensions) in the data while keeping most of the important information. Here's how PCA works in simple terms


* 1. Understanding Dimensions:
 Imagine a dataset with many features (e.g., height, weight, age, income). Each feature represents a dimension. Too many dimensions make analysis difficult and can lead to overfitting in models.

* 2. Finding Principal Components:
PCA identifies new features called principal components, which are combinations of the original features.
The first principal component captures the most variation (spread) in the data.
The second principal component captures the next most variation, and so on. Importantly, these components are uncorrelated (independent).
Transforming the Data:

* 3. The dataset is rotated and transformed onto the new principal components.
You can choose to keep only a few of these components (e.g., the top 2 or 3) that explain most of the data's variation.
Why Use PCA?:

* 4. Reduce Complexity: Simplifies large datasets by keeping only the most informative components.
Remove Redundancy: Gets rid of overlapping information between features.
* 5. Visualization: Helps visualize high-dimensional data in 2D or 3D plots.
Improves Models: Reducing dimensions can speed up computations and improve model performance.

Key Takeaways:
PCA is a tool for reducing dimensions without losing much important information.
It finds new variables (principal components) that summarize the data's structure.
Often used in exploratory data analysis, machine learning, and data compression.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
df.describe()
```

### As we've noticed before it is dificult to visualize the hight dimensional data, we can use PCA to find the first tow principal components, and visualize the data in this new 2-D space, with a single scatter plot. Before do this though, we'll need to scale our data so that each feature has a aingle unit variance.

```python
df.shape
```

```python
Scalar = StandardScaler()
df_scaled = Scalar.fit_transform(df)
```

```python
# PCA
pca = PCA(n_components=2)
pca.fit(df_scaled)
```

```python
x_pca = pca.transform(df_scaled)
x_pca.shape
```

```python
x_pca
```

```python
# we have reduced 30 demensions in to two lets plot this two
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap= 'plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA')
plt.colorbar()
plt.legend= 'True'
plt.show()
```

```python
# ploting 3d
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_title('PCA')
plt.show()
```
